## Tasks

Now write and execute SQL queries to solve the assignment tasks.

### Task 1

##### Display the names of the unique launch sites  in the space mission


```python
df_unique_launchsites=pd.read_sql_query("Select distinct Launch_Site  from spacex_v5 ",conn) 
print(df_unique_launchsites)
```

### Task 2

##### Display 5 records where launch sites begin with the string 'CCA'


```python
#%sql SELECT Launch_Site from spacex_v2 where (Launch_Site) LIKE 'CCA%' LIMIT 5;
df_launchsites_CCA5=pd.read_sql_query("Select Launch_Site from spacex_v5 where Launch_Site Like 'CCA%' Limit 5",conn) 
print(df_launchsites_CCA5)
```

### Task 3

##### Display the total payload mass carried by boosters launched by NASA (CRS)


```python
#  For validation purposes... sum in df_NASA_CRS  'PAYLOAD_MASS__KG_'   column 

df_NASA_CRS=pd.read_sql_query("Select * from spacex_v5  where Customer='NASA (CRS)'",conn) 

print(df_NASA_CRS.head(2))
#df_NASA_CRS['PAYLOAD_MASS__KG_'].sum()
print('----')
print('Total payload mass, customer= NASA (CRS):', df_NASA_CRS['PAYLOAD_MASS__KG_'].sum())
```

```python
#  Based on SQL only...

payload_NASA_CRS=pd.read_sql_query("Select sum(PAYLOAD_MASS__KG_) as 'Total payload mass NASA CRS' from spacex_v5 where Customer='NASA (CRS)'",conn) 
print(payload_NASA_CRS)
```

### Task 4

##### Display average payload mass carried by booster version F9 v1.1


```python
payload_F9v11=pd.read_sql_query("Select avg(PAYLOAD_MASS__KG_) as 'avg mass' from spacex_v5 where Booster_Version='F9 v1.1'",conn) 
print(payload_F9v11)
```

### Task 5

##### List the date when the first successful landing outcome in ground pad was acheived.

*Hint:Use min function*


```python
df_unique_launchsites=pd.read_sql_query("Select distinct Launch_Site  from spacex_v5 ",conn) 
print(df_unique_launchsites)
```

```python
df_unique_Landing_Outcome=pd.read_sql_query("Select distinct Landing_Outcome from spacex_v5 ",conn) 
print(df_unique_Landing_Outcome)
```

```python
min_Date_success_landing=pd.read_sql_query("select min(Date) from spacex_v5 where Landing_Outcome = 'Success (ground pad)'",conn) 
print(min_Date_success_landing)
```

### Task 6

##### List the names of the boosters which have success in drone ship and have payload mass greater than 4000 but less than 6000


```python
#  sql query 
q_boost_succ= """ select  Booster_Version from spacex_v5 where Landing_Outcome = 'Success (drone ship)' 
                    and PAYLOAD_MASS__KG_ > 4000 
                    and PAYLOAD_MASS__KG_ < 6000 """
```

```python
Booster_success_landing=pd.read_sql_query(q_boost_succ,conn)

print(Booster_success_landing)
```

### Task 7

##### List the total number of successful and failure mission outcomes


```python
#  sql query 
qsf= """Select (Select Count(Mission_Outcome) from spacex_v5 where Mission_Outcome like '%Success%')
        as Successful_Missions,
        (Select Count(Mission_Outcome) from spacex_v5
        where Mission_Outcome like '%Failure%') as Failed_Missions """


success_failure= pd.read_sql_query(qsf,conn) 
print(success_failure)
```

### Task 8

##### List the   names of the booster_versions which have carried the maximum payload mass. Use a subquery


```python
# preliminary query on disctinct booster versions  
count_booster= pd.read_sql_query("Select distinct Booster_Version from spacex_v5",conn) 
print(count_booster)
```

```python
#  sql query 
qboost= """Select distinct Booster_Version, max(PAYLOAD_MASS__KG_) as max_payload_mass 
from spacex_v5
group by Booster_Version 
order by max_payload_mass  desc"""
```

```python
boost_max_load= pd.read_sql_query(qboost,conn) 
boost_max_load.head(5)
```

### Task 9

##### List the failed landing_outcomes in drone ship, their booster versions, and launch site names for in year 2015


```python
df.head(2)
```

```python
#  sql query 
q_failed_landing= """ Select Date, Booster_Version, Launch_Site, Landing_Outcome 
                   from spacex_v5 
                   where Landing_Outcome = 'Failure (drone ship)'
                   and Date like '%2015%'   """

```

```python
fail_drone= pd.read_sql_query(q_failed_landing,conn) 
fail_drone.head(5)
```

### Task 10

##### Rank the count of landing outcomes (such as Failure (drone ship) or Success (ground pad)) between the date 2010-06-04 and 2017-03-20, in descending order


```python
#  sql query 
q_count_landing= """ Select Landing_Outcome, count(*) as count_landings  
                   from spacex_v5                
                   where Date between '2010-06-04' and '2017-03-20'
                   group by Landing_Outcome
                   order by count_landings desc  """
```

```python
count_landing= pd.read_sql_query(q_count_landing,conn) 
count_landing.head(10)
```
